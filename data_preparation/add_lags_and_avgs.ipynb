```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
multi = pd.read_csv('../data/regional_multivariate_data.csv')
multi['region'].unique()
```

```python
enc_data = multi[multi['region'] == "East North Central"]
esc_data = multi[multi['region'] == "East South Central"]
mid_atlantic_data = multi[multi['region'] == "Mid-Atlantic"]
mountain_data = multi[multi['region'] == "Mountain"]
new_england_data = multi[multi['region'] == "New England"]
pacific_data = multi[multi['region'] == "Pacific"]
south_atlantic_data = multi[multi['region'] == "South Atlantic"]
wnc_data = multi[multi['region'] == "West North Central"]
wsc_data = multi[multi['region'] == "West South Central"]
```

```python
pacific_lag = [1, 3, 7, 10]
enc_lag = [1, 2, 7]
esc_lag = [1, 2, 3, 7]
mid_atlantic_lag = [1, 2, 8, 9]
mountain_lag = [1, 6, 7]
new_england_lag = [1, 7]
south_atlantic_lag = [1, 6, 7]
wnc_lag = [1, 6, 7]
wsc_lag = [1, 7]

for i in pacific_lag:
    pacific_data[str(i) + "_day_shift"] = pacific_data['new_confirmed'].shift(i)
    pacific_data["7_day_avg"] = pacific_data['new_confirmed'].rolling(window= 7).mean()

for i in enc_lag:
    enc_data[str(i) + "_day_shift"] = enc_data['new_confirmed'].shift(i)
    enc_data["7_day_avg"] = enc_data['new_confirmed'].rolling(window= 7).mean()

for i in esc_lag:
    esc_data[str(i) + "_day_shift"] = esc_data['new_confirmed'].shift(i)
    esc_data["7_day_avg"] = esc_data['new_confirmed'].rolling(window= 7).mean()

for i in mid_atlantic_lag:
    mid_atlantic_data[str(i) + "_day_shift"] = mid_atlantic_data['new_confirmed'].shift(i)
    mid_atlantic_data["7_day_avg"] = mid_atlantic_data['new_confirmed'].rolling(window= 7).mean()

for i in mountain_lag:
    mountain_data[str(i) + "_day_shift"] = mountain_data['new_confirmed'].shift(i)
    mountain_data["7_day_avg"] = mountain_data['new_confirmed'].rolling(window= 7).mean()

for i in new_england_lag:
    new_england_data[str(i) + "_day_shift"] = new_england_data['new_confirmed'].shift(i)
    new_england_data["7_day_avg"] = new_england_data['new_confirmed'].rolling(window= 7).mean()

for i in south_atlantic_lag:
    south_atlantic_data[str(i) + "_day_shift"] = south_atlantic_data['new_confirmed'].shift(i)
    south_atlantic_data["7_day_avg"] = south_atlantic_data['new_confirmed'].rolling(window= 7).mean()

for i in wnc_lag:
    wnc_data[str(i) + "_day_shift"] = wnc_data['new_confirmed'].shift(i)
    wnc_data["7_day_avg"] = wnc_data['new_confirmed'].rolling(window= 7).mean()

for i in wsc_lag:
    wsc_data[str(i) + "_day_shift"] = wsc_data['new_confirmed'].shift(i)
    wsc_data["7_day_avg"] = wsc_data['new_confirmed'].rolling(window= 7).mean()
```

```python
pacific_data.to_csv('../data/regional_datasets/pacific_data.csv')
enc_data.to_csv('../data/regional_datasets/enc_data.csv')
esc_data.to_csv('../data/regional_datasets/esc_data.csv')
mid_atlantic_data.to_csv('../data/regional_datasets/mid_atlantic_data.csv')
mountain_data.to_csv('../data/regional_datasets/mountain_data.csv')
new_england_data.to_csv('../data/regional_datasets/new_england_data.csv')
south_atlantic_data.to_csv('../data/regional_datasets/south_atlantic_data.csv')
wnc_data.to_csv('../data/regional_datasets/wnc_data.csv')
wsc_data.to_csv('../data/regional_datasets/wsc_data.csv')
```

```python
plt.rcParams['figure.figsize'] = [15, 6]

wsc_data.plot(x='date', y=['new_confirmed', '7_day_avg'])

plt.title('New Confirmed Cases and 7-Day Average', fontsize=20)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Cases', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=14)

plt.show()
```
