```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
multi = pd.read_csv('../data/regional_multivariate_data.csv')
univ = pd.read_csv('../data/regional_univariate_data.csv')
multi['region'].unique()
```

```python
# Multivariate
enc_data_multi = multi[multi['region'] == "East North Central"]
esc_data_multi = multi[multi['region'] == "East South Central"]
mid_atlantic_data_multi = multi[multi['region'] == "Mid-Atlantic"]
mountain_data_multi = multi[multi['region'] == "Mountain"]
new_england_data_multi = multi[multi['region'] == "New England"]
pacific_data_multi = multi[multi['region'] == "Pacific"]
south_atlantic_data_multi = multi[multi['region'] == "South Atlantic"]
wnc_data_multi = multi[multi['region'] == "West North Central"]
wsc_data_multi = multi[multi['region'] == "West South Central"]

# Univariate
enc_data_univ = univ[univ['region'] == "East North Central"]
esc_data_univ = univ[univ['region'] == "East South Central"]
mid_atlantic_data_univ = univ[univ['region'] == "Mid-Atlantic"]
mountain_data_univ = univ[univ['region'] == "Mountain"]
new_england_data_univ = univ[univ['region'] == "New England"]
pacific_data_univ = univ[univ['region'] == "Pacific"]
south_atlantic_data_univ = univ[univ['region'] == "South Atlantic"]
wnc_data_univ = univ[univ['region'] == "West North Central"]
wsc_data_univ = univ[univ['region'] == "West South Central"]
```

```python
pacific_lag = [1, 3, 7, 10]
enc_lag = [1, 2, 7]
esc_lag = [1, 2, 3, 7]
mid_atlantic_lag = [1, 2, 8, 9]
mountain_lag = [1, 6, 7]
new_england_lag = [1, 7]
south_atlantic_lag = [1, 6, 7]
wnc_lag = [1, 6, 7]
wsc_lag = [1, 7]

for i in pacific_lag:
    pacific_data_multi[str(i) + "_day_shift"] = pacific_data_multi['new_confirmed'].shift(i)
    pacific_data_multi["7_day_avg"] = pacific_data_multi['new_confirmed'].rolling(window= 7).mean()

    pacific_data_univ[str(i) + "_day_shift"] = pacific_data_univ['new_confirmed'].shift(i)
    pacific_data_univ["7_day_avg"] = pacific_data_univ['new_confirmed'].rolling(window= 7).mean()

for i in enc_lag:
    enc_data_multi[str(i) + "_day_shift"] = enc_data_multi['new_confirmed'].shift(i)
    enc_data_multi["7_day_avg"] = enc_data_multi['new_confirmed'].rolling(window= 7).mean()

    enc_data_univ[str(i) + "_day_shift"] = enc_data_univ['new_confirmed'].shift(i)
    enc_data_univ["7_day_avg"] = enc_data_univ['new_confirmed'].rolling(window= 7).mean()

for i in esc_lag:
    esc_data_multi[str(i) + "_day_shift"] = esc_data_multi['new_confirmed'].shift(i)
    esc_data_multi["7_day_avg"] = esc_data_multi['new_confirmed'].rolling(window= 7).mean()

    esc_data_univ[str(i) + "_day_shift"] = esc_data_univ['new_confirmed'].shift(i)
    esc_data_univ["7_day_avg"] = esc_data_univ['new_confirmed'].rolling(window= 7).mean()


for i in mid_atlantic_lag:
    mid_atlantic_data_multi[str(i) + "_day_shift"] = mid_atlantic_data_multi['new_confirmed'].shift(i)
    mid_atlantic_data_multi["7_day_avg"] = mid_atlantic_data_multi['new_confirmed'].rolling(window= 7).mean()

    mid_atlantic_data_univ[str(i) + "_day_shift"] = mid_atlantic_data_univ['new_confirmed'].shift(i)
    mid_atlantic_data_univ["7_day_avg"] = mid_atlantic_data_univ['new_confirmed'].rolling(window= 7).mean()

for i in mountain_lag:
    mountain_data_multi[str(i) + "_day_shift"] = mountain_data_multi['new_confirmed'].shift(i)
    mountain_data_multi["7_day_avg"] = mountain_data_multi['new_confirmed'].rolling(window= 7).mean()

    mountain_data_univ[str(i) + "_day_shift"] = mountain_data_univ['new_confirmed'].shift(i)
    mountain_data_univ["7_day_avg"] = mountain_data_univ['new_confirmed'].rolling(window= 7).mean()

for i in new_england_lag:
    new_england_data_multi[str(i) + "_day_shift"] = new_england_data_multi['new_confirmed'].shift(i)
    new_england_data_multi["7_day_avg"] = new_england_data_multi['new_confirmed'].rolling(window= 7).mean()

    new_england_data_univ[str(i) + "_day_shift"] = new_england_data_univ['new_confirmed'].shift(i)
    new_england_data_univ["7_day_avg"] = new_england_data_univ['new_confirmed'].rolling(window= 7).mean()

for i in south_atlantic_lag:
    south_atlantic_data_multi[str(i) + "_day_shift"] = south_atlantic_data_multi['new_confirmed'].shift(i)
    south_atlantic_data_multi["7_day_avg"] = south_atlantic_data_multi['new_confirmed'].rolling(window= 7).mean()

    south_atlantic_data_univ[str(i) + "_day_shift"] = south_atlantic_data_univ['new_confirmed'].shift(i)
    south_atlantic_data_univ["7_day_avg"] = south_atlantic_data_univ['new_confirmed'].rolling(window= 7).mean()


for i in wnc_lag:
    wnc_data_multi[str(i) + "_day_shift"] = wnc_data_multi['new_confirmed'].shift(i)
    wnc_data_multi["7_day_avg"] = wnc_data_multi['new_confirmed'].rolling(window= 7).mean()

    wnc_data_univ[str(i) + "_day_shift"] = wnc_data_univ['new_confirmed'].shift(i)
    wnc_data_univ["7_day_avg"] = wnc_data_univ['new_confirmed'].rolling(window= 7).mean()

for i in wsc_lag:
    wsc_data_multi[str(i) + "_day_shift"] = wsc_data_multi['new_confirmed'].shift(i)
    wsc_data_multi["7_day_avg"] = wsc_data_multi['new_confirmed'].rolling(window= 7).mean()

    wsc_data_univ[str(i) + "_day_shift"] = wsc_data_univ['new_confirmed'].shift(i)
    wsc_data_univ["7_day_avg"] = wsc_data_univ['new_confirmed'].rolling(window= 7).mean()
```

```python
pacific_data_multi.to_csv('../data/regional_datasets/multivariate/pacific_data_multi.csv')
enc_data_multi.to_csv('../data/regional_datasets/multivariate/enc_data_multi.csv')
esc_data_multi.to_csv('../data/regional_datasets/multivariate/esc_data_multi.csv')
mid_atlantic_data_multi.to_csv('../data/regional_datasets/multivariate/mid_atlantic_data_multi.csv')
mountain_data_multi.to_csv('../data/regional_datasets/multivariate/mountain_data_multi.csv')
new_england_data_multi.to_csv('../data/regional_datasets/multivariate/new_england_data_multi.csv')
south_atlantic_data_multi.to_csv('../data/regional_datasets/multivariate/south_atlantic_data_multi.csv')
wnc_data_multi.to_csv('../data/regional_datasets/multivariate/wnc_data_multi.csv')
wsc_data_multi.to_csv('../data/regional_datasets/multivariate/wsc_data_multi.csv')

pacific_data_univ.to_csv('../data/regional_datasets/univariate/pacific_data_univ.csv')
enc_data_univ.to_csv('../data/regional_datasets/univariate/enc_data_univ.csv')
esc_data_univ.to_csv('../data/regional_datasets/univariate/esc_data_univ.csv')
mid_atlantic_data_univ.to_csv('../data/regional_datasets/univariate/mid_atlantic_data_univ.csv')
mountain_data_univ.to_csv('../data/regional_datasets/univariate/mountain_data_univ.csv')
new_england_data_univ.to_csv('../data/regional_datasets/univariate/new_england_data_univ.csv')
south_atlantic_data_univ.to_csv('../data/regional_datasets/univariate/south_atlantic_data_univ.csv')
wnc_data_univ.to_csv('../data/regional_datasets/univariate/wnc_data_univ.csv')
wsc_data_univ.to_csv('../data/regional_datasets/univariate/wsc_data_univ.csv')
```

```python
plt.rcParams['figure.figsize'] = [15, 6]

wsc_data_multi.plot(x='date', y=['new_confirmed', '7_day_avg'])

plt.title('New Confirmed Cases and 7-Day Average', fontsize=20)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Cases', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=14)

plt.show()
```
